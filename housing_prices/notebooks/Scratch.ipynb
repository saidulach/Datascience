```python
feature_description = {}
#print(type(feature_description))
with open ('../data/raw/data_description.txt', 'r') as file:
    for line in file:
        if ":" in line:
            key, value = line.strip().split(':',1)
            feature_description[key.strip()] = value.strip()

df = pd.Series(feature_description, name = 'Description').to_frame()
df.head()

df.to_csv('features.csv', index_label = 'Feature')
```
